```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('train_loss.pickle', mode='rb') as f:
    train_loss = pickle.load(f)
```

```python
train_loss
```

```python
with open('valid_loss.pickle', mode='rb') as f:
    valid_loss = pickle.load(f)
```

```python
valid_loss
```

```python
len(train_loss)
```

```python
import matplotlib.pyplot as plt

# visualize the loss as the network trained
fig = plt.figure(figsize=(10,8))
plt.plot(range(1,len(train_loss)+1),train_loss, label='Training Loss')
plt.plot(range(1,len(valid_loss)+1),valid_loss,label='Validation Loss')

# find position of lowest validation loss
minposs = valid_loss.index(min(valid_loss))+1 
plt.axvline(minposs, linestyle='--', color='r',label='Early Stopping Checkpoint')

plt.xlabel('epochs')
plt.ylabel('loss')
plt.ylim(0, 3) # consistent scale
plt.xlim(1, len(train_loss)+1) # consistent scale
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
fig.savefig('loss_plot_epoc5.png', bbox_inches='tight')
```
